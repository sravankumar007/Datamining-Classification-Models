# Classification - Nearest Neighbors

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_data = load_iris() #loading the iris data
```

```python
knn_data = pd.DataFrame(data= np.c_[iris_data['data'], iris_data['target']], columns= iris_data['feature_names'] + ['target']) #converting the data to dataframe
```

```python
print(knn_data)
```

```python
#Taking only two attributes/features for calculating euclidean distance
X = knn_data[['sepal length (cm)', 'sepal width (cm)']]
Y = knn_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
#training the model with 70 percent of the data
#assigning 30 percent of data for testing
```

```python
classifier = KNeighborsClassifier(n_neighbors=1, metric='euclidean') #using metric as euclidean as mentioned in the given criteria
classifier.fit(X_train, Y_train) #training the model
```

```python
Y_pred = classifier.predict(X_test) #predecting the values for the test data
```

```python
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))

plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=Y_pred,cmap='coolwarm')
#The below graph shows the target value for the taken x=sepal length (cm) , y=sepal width (cm) values
```

```python
classifier1 = KNeighborsClassifier(n_neighbors=4, metric='euclidean') #using metric as euclidean as mentioned in the given criteria
classifier1.fit(X_train, Y_train)
Y_pred = classifier1.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))

plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=Y_pred, cmap='coolwarm')
#The below graph shows the target value for the taken x=sepal length (cm) , y=sepal width (cm) values
```

```python
classifier2 = KNeighborsClassifier(n_neighbors=11, metric='euclidean') #using metric as euclidean as mentioned in the given criteria
classifier2.fit(X_train, Y_train)
Y_pred = classifier2.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))

plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=Y_pred, cmap='coolwarm')
#The below graph shows the target value for the taken x=sepal length (cm) , y=sepal width (cm) values
```
